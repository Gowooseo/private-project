3.LSTM

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
```

```python


# 데이터 준비
cols = moneyDFT_Number.columns[0:]
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(moneyDFT_Number[cols])

# 데이터 분할
train_size = int(len(scaled_data) * 0.8)
train_data = scaled_data[:train_size]
test_data = scaled_data[train_size:]

# 데이터 전처리
look_back = 30  # 이전 30일 데이터 사용
X_train, y_train = [], []
for i in range(look_back, len(train_data)):
    X_train.append(train_data[i-look_back:i])
    y_train.append(train_data[i])
X_train, y_train = np.array(X_train), np.array(y_train)

# LSTM 모델 구축 및 학습
model = Sequential()
model.add(LSTM(64, input_shape=(look_back, len(cols))))
model.add(Dense(len(cols)))
model.compile(optimizer='adam', loss='mse')
model.fit(X_train, y_train, epochs=100, batch_size=32)

# 예측
X_test = []
for i in range(look_back, len(scaled_data)):
    X_test.append(scaled_data[i-look_back:i])
X_test = np.array(X_test)
predictions = model.predict(X_test)

# 예측 값 역정규화
predictions = scaler.inverse_transform(predictions)


# # 예측 결과 출력
# future_dates = pd.date_range(start=moneyDFT_Number['index'].max(), periods=len(predictions), freq='D')
# for i, col in enumerate(cols):
#     col_name = f"{col}_pred"  # 열 이름 변경
#     print(f"{col} 예측 결과:")
#     result = pd.DataFrame({col_name: predictions[:, i]}, index=future_dates)
#     print(result.tail(30))

# 예측 결과 출력 및 특정 날짜의 환율 출력
def print_prediction_results(predictions, dates, target_date, target_currency):
    target_index = dates.get_loc(target_date)
    print(f"{target_date}의 {target_currency} 예측 결과: {predictions[target_index]:.2f}")

# 사용자 입력 날짜
target_date = input("확인하고 싶은 날짜를 입력하세요 (YYYY-MM-DD 형식): ")

# 예측 결과 출력
future_dates = pd.date_range(start=moneyDFT_Number.index.max(), periods=len(predictions), freq='D')
for col in moneyDFT_Number.columns[:]:
    col_name = f"{col}_pred"  # 열 이름 변경
    if col == '원/미국달러(매매기준율)':
        col_name = '원/달러 환율'
    print(f"{col} 예측 결과:")
    result = pd.DataFrame({col_name: predictions[:, moneyDFT_Number.columns.get_loc(col)-1]}, index=future_dates)
    print(result.to_string())
    print()

# 특정 날짜의 환율 출력
print_prediction_results(predictions[:, moneyDFT_Number.columns.get_loc(target_currency)-1], future_dates, target_date, target_currency)


```
